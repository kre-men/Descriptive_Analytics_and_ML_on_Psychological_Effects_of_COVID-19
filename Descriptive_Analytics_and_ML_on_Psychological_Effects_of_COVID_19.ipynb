```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import learning_curve
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
X = num_data_class.drop('self_time', axis=1)
y = num_data_class['self_time']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

poly = PolynomialFeatures(degree=2)

poly_features_train = poly.fit_transform(X_train_scaled)
poly_features_test = poly.transform(X_test_scaled)

model = LinearRegression(fit_intercept=True)
model.fit(poly_features_train, y_train)

predictions_train = model.predict(poly_features_train)
predictions_test = model.predict(poly_features_test)

MAE_train = mean_absolute_error(y_train, predictions_train)
RMSE_train = np.sqrt(mean_squared_error(y_train, predictions_train))

MAE_test = mean_absolute_error(y_test, predictions_test)
RMSE_test = np.sqrt(mean_squared_error(y_test, predictions_test))
```

```python
print("Training Set Metrics:")
print("MAE: %f" % MAE_train)
print("RMSE: %f" % RMSE_train)

print("\nTest Set Metrics:")
print("MAE: %f" % MAE_test)
print("RMSE: %f" % RMSE_test)
```

```python
pipeline = Pipeline([
    ('poly_features', PolynomialFeatures()),
    ('model', LinearRegression())
])

param_grid = {'poly_features__degree': [1, 2, 3, 4, 5]}

grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X, y)

best_degree = grid_search.best_params_['poly_features__degree']
print(f'Best Degree for Polynomial Regression: {best_degree}')
```

```python
scaler = MinMaxScaler()
columns_to_normalize = ['prod_inc', 'sleep_bal', 'new_skill', 'fam_connect', 'relaxed', 'self_time']
scaled_data = num_data_class.copy()
scaled_data[columns_to_normalize] = scaler.fit_transform(scaled_data[columns_to_normalize])

X = scaled_data.drop('self_time', axis=1)
y = scaled_data['self_time']

degree = 1
poly = PolynomialFeatures(degree=degree)
poly_features = poly.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.2, random_state=0)

model = LinearRegression(fit_intercept=True)
model.fit(X_train, y_train)

predictions_train = model.predict(X_train)
predictions_test = model.predict(X_test)

plt.scatter(y_train, predictions_train, color='blue', label='Training Set')
plt.scatter(y_test, predictions_test, color='green', label='Test Set', alpha=0.5)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs. Predicted Values")
plt.legend()
plt.show()

MAE_train = mean_absolute_error(y_train, predictions_train)
RMSE_train = np.sqrt(mean_squared_error(y_train, predictions_train))

MAE_test = mean_absolute_error(y_test, predictions_test)
RMSE_test = np.sqrt(mean_squared_error(y_test, predictions_test))

MSE_train = mean_squared_error(y_train, predictions_train)
MSE_test = mean_squared_error(y_test, predictions_test)

R2_train = r2_score(y_train, predictions_train)
R2_test = r2_score(y_test, predictions_test)

comp_table = pd.DataFrame({'Training Set Metrics':
                               {'MAE':MAE_train,
                                'RMSE':RMSE_train,
                                'R2':R2_train,
                                'MSE':MSE_train},
                           'Test Set Metrics':
                               {'MAE':MAE_test,
                                'RMSE':RMSE_test,
                                'R2':R2_test,
                                'MSE':MSE_test}})

print(f"Value of 'Degree': {degree}")
print(comp_table)
```

```python
from kmodes.kmodes import KModes
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from collections import Counter
from sklearn import metrics
```

```python
# scaled_data_clust = pd.read_csv('scaled_data_clust.csv')
```

```python
# label_data = pd.read_csv('label_data.csv')
```

```python
cost = []
for num_clusters in range(1, 7):
    kmode = KModes(n_clusters=num_clusters, init='random', n_init=5, verbose=0, random_state=0)
    kmode.fit_predict(label_data)
    cost.append(kmode.cost_)

plt.plot(range(1, 7), cost, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Cost')
plt.show()
```

```python
silhouette_scores = []

for k in range(2, 11):
    km = KModes(n_clusters=k, init='random', n_init=5, verbose=0, random_state=0)
    clusters = km.fit_predict(label_data)
    silhouette_scores.append(metrics.silhouette_score(label_data, clusters))

plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Different K Values')
plt.show()
```

```python
num_clusts = 4
kmode = KModes(n_clusters=num_clusts, init='random', n_init=5, verbose=0, random_state=0)
cluster_labels = kmode.fit_predict(label_data)
```

```python
silhouette_avg = silhouette_score(label_data, cluster_labels)
print(f"Silhouette Score: {silhouette_avg}")
```

```python
cluster_counts = Counter(cluster_labels)
```

```python
for cluster, count in cluster_counts.items():
    print(f'Cluster {cluster + 1}: {count} instances')
```

```python
percentage_per_cluster = [count / len(cluster_labels) * 100 for count in cluster_counts.values()]
for cluster, percentage in zip(cluster_counts.keys(), percentage_per_cluster):
    print(f'Cluster {cluster + 1}: {percentage:.2f}% of instances')
```

```python
X = label_data.columns

for i in range(num_clusts):
    cluster_mean = label_data[cluster_labels == i].mean()
    plt.barh(X, cluster_mean, label=f'Cluster {i + 1}')

plt.ylabel("Features")
plt.xlabel("Mean values")
plt.title("Mean values per attribute for each cluster")
plt.legend()
plt.show()
```
